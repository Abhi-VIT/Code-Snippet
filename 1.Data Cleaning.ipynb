# **1. Generate Synthetic Stock Price Data**

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.seed(1)
days = np.arange(1, 101)
prices = 100 + np.cumsum(np.random.normal(0,1, size = 100))
df = pd.DataFrame({'Day': days, 'Stock_Prices': prices})
```

# **3. Rolling Average Calculation**
##### **o Add a new column Rolling_Mean_5 (5-day moving average)**
##### **o Add another column Rolling_Std_5 (5-day rolling standard deviation)**

```python
s = pd.Series(df.Stock_Prices)
```

```python
df['Rolling_Mean_5'] = s.rolling(window=6).mean()
```

```python
df.head(50)
```

```python
p = pd.Series(df.Stock_Prices)
```

```python
df['Rolling_Std_5'] = p.rolling(window=6).std()
```

```python
df.head(50)
```

# **Objective:**

```python
age = np.random.randint(20, 70, size = 200)
```

```python
age
```

```python
weight = np.random.normal(65, 12, size = 200)
```

```python
height = np.random.normal(165, 10, size = 200)
```

```python
cholestrol = np.random.uniform(150.0, 300.0, size = 200)
```

```python
noise = np.random.normal(0,1,200)
```

```python
blood_pressure = [(80 + 0.5*age+ n) for age , n in zip(age, noise)]
```

```python
data = pd.DataFrame({'age' : age,
                     'weight(kg)' : weight,
                     'height(cm)' : height, 
                     'cholesterol' : cholestrol,
                     'Blood_Pressure' : blood_pressure})
data
```

# **Introduce 5% mising values in weight**

```python
missing_percentage = 0.05
num_missing = int(len(data) * missing_percentage)
missing_indices = np.random.choice(data.index, num_missing, replace=False)
data.loc[missing_indices, 'weight(kg)'] = np.nan
```

# **2. Data Cleaning**
##### **o Identify and handle missing values (impute with mean or median)**
##### **o Detect outliers using:**
##### ** IQR method**
##### ** Z-score method (NumPy)**

```python
data.isnull().sum()
```

```python
mean_value = data['weight(kg)'].mean()
data['weight(kg)'].fillna(value = mean_value, inplace = True)
mean_cho = data['cholesterol'].median()
data['cholesterol'].fillna(value= mean_cho, inplace = True)
```

```python
data.isnull().sum()
```

```python
sns.lineplot(data)
```

```python
sns.boxplot(data)
```

```python
sns.boxplot(data['cholesterol'])
```

```python
sns.boxplot(data['Blood_Pressure'])
```

##### **Z - Score**

```python
from scipy.stats import zscore
```

```python
data1 = data.copy()
```

```python
data1['zscore_cho'] = zscore(data.cholesterol)
data1['zscore_blood'] = zscore(data.Blood_Pressure)
```

```python
data1
```

```python
outliers_cho = data1[data1.zscore_cho.abs() > 2]
```

```python
print(outliers_cho)
```

```python
outliers_blood = data1[data1.zscore_blood.abs() > 2]
print(outliers_blood)
```
